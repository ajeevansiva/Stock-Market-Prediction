# **Part A – Application area review**

Introduction:

In the dynamic realm of financial markets, accurate prediction of stock prices remains a complex challenge, influenced by numerous variables and unpredictable events. In response to this challenge, the integration of artificial intelligence (AI) has emerged as a transformative approach. AI leverages advanced algorithms, including machine learning and deep learning models, to analyze vast datasets and discern intricate patterns within stock market movements. This technological infusion aims to enhance predictive accuracy, facilitate real-time decision-making, and contribute to effective risk mitigation in portfolio management. However, amidst the potential benefits, a persistent problem lies in the inherent volatility and unpredictability of financial markets, necessitating robust AI solutions to navigate and capitalize on market complexities while minimizing risks for investors and traders (Patel et al., 2021)

**Problem Domain**

Algorithmic trading, a systematic approach to buying or selling in financial markets, has seen significant advancements through the integration of Artificial Intelligence (AI) techniques. AI, including machine learning, deep learning, and natural language processing, has been applied to enhance decision-making and improve trading accuracy and efficiency (Khare et al., 2017). AI-driven trading systems can process and analyze large volumes of data in real-time, leading to more precise trading decisions and reduced human involvement, thereby improving speed and efficiency (Bathla, 2020). High-frequency trading (HFT) algorithms, powered by AI, execute thousands of trades per second, leveraging complex mathematical models and real-time data analysis. These algorithms respond to market conditions at speeds human traders can never match, profiting from tiny price differentials.AI has also been used to enhance risk management in algorithmic trading, as AI-driven trading systems can adapt and learn from market data to enhance decision-making(Baihaqi et al., 2023). The use of AI in algorithmic trading has revolutionized financial markets, offering benefits such as fast trading speeds, improved accuracy, and enhanced risk management. However, it is important to note that AI trading strategies are not without risks, and it's crucial to evaluate the performance and track record of any system before using it for trading.The integration of AI in algorithmic trading has the potential to transform trading processes, but it is essential to proceed with caution and conduct proper due diligence when using AI in trading (Rahman et al., 2021).

**Existing Works**

Recent advances in Artificial Intelligence (AI) have made algorithmic trading play a central role in finance. However, current research and applications are disconnected information islands. A generally applicable pipeline for designing, programming, and evaluating the algorithmic trading of stock and crypto assets has been proposed. Another innovative approach based on deep reinforcement learning (DRL) to solve the algorithmic trading problem of determining the optimal trading position at any point in time during a trading activity in the stock market has been proposed (Theate and Ernst, no date).

In conclusion, Algorithmic Trading has become a significant part of modern financial markets, driven by the desire for cost and time savings. However, it presents several challenges, particularly at the strategy level and in terms of risk management and controls. The application of AI in this domain has shown promising results, with research and applications offering systematic ways to program, evaluate, and compare different trading strategies. Despite the progress, the field is still characterized by disconnected information islands, indicating the need for further research and development. The AI Techniques used in Algorithmic Trading


1. Machine Learning (ML) in Algorithmic Trading:

Machine learning techniques are extensively applied in algorithmic trading to analyze market data, identify patterns, and make predictions for optimal trading decisions. Supervised learning models, such as Support Vector Machines or Random Forests, can be trained on historical market data to recognize patterns associated with price movements or specific trading signals. Feature engineering is crucial in this context, where relevant market indicators, technical analysis metrics, and economic indicators serve as input features. The trained model can then be deployed to make real-time trading decisions based on incoming market data, helping to automate trading strategies.


2. Deep Learning (DL) in Algorithmic Trading:

Deep learning, particularly neural networks, is increasingly employed in algorithmic trading for its ability to capture intricate patterns in large and complex datasets. Recurrent Neural Networks (RNNs) and Long Short-Term Memory (LSTM) networks are well-suited for time-series data in financial markets. These models can learn temporal dependencies and non-linear relationships in market movements, enabling the prediction of future price trends. Deep reinforcement learning is another approach, where neural networks are trained to optimize trading strategies by interacting with simulated or historical market data. The adaptability and capacity to handle vast amounts of data make deep learning a powerful tool in developing sophisticated algorithmic trading systems.

3. Natural Language Processing (NLP) in Algorithmic Trading:

Natural Language Processing plays a crucial role in algorithmic trading by extracting valuable insights from financial news, social media, and other textual sources. Sentiment analysis, a subfield of NLP, is employed to gauge the sentiment surrounding financial instruments. News articles, tweets, and financial reports can be analyzed to assess market sentiment, and this information can be incorporated into trading strategies. Named Entity Recognition (NER) helps identify key entities mentioned in news articles, such as companies or economic indicators, influencing trading decisions. NLP can be combined with other techniques, like ML or DL, to create comprehensive trading systems that consider both quantitative market data and qualitative textual information, enhancing the adaptability of algorithmic trading strategies.

```python
import pandas as pd
import pandas_datareader as pdr
import yfinance as yf
yf.pdr_override()
```

```python
import datetime
import time

end = datetime.datetime(2023, 12, 30)
start = datetime.datetime(2011, 1, 1)

# Get the stock data from Yahoo Finance
df = yf.download(tickers=['TSLA'], start=start, end=end)
df.head()
```

```python
df.tail()
```

```python
# Checking if there's any empty or null values
df.info()
```

```python
# Drop None or Null values
df = df.dropna()
```

```python
# generating descriptive statistics
df.describe()
```

```python
#Plotting close price of the Tesla Stock

from matplotlib import pyplot as plt
df['Close'].plot(kind='line', figsize=(12, 6), title='Close Price')
plt.gca().spines[['top', 'right']].set_visible(False)
```

The stock prices are highly volatile and prices change quickly with time. To observe any trend or pattern we can take the help of a 30-day 200-day average.

```python
# Calculate the 30-day and 100-day moving averages
df['MA30'] = df['Close'].rolling(window=30).mean()
df['MA100'] = df['Close'].rolling(window=100).mean()

plt.figure(figsize=(12, 6))
# 30-day moving average line plot
df['MA30'].plot(kind='line', label='30-day MA')

# 100-day moving average line plot
df['MA100'].plot(kind='line', label='100-day MA')

plt.title('Moving Averages')
plt.legend()

plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

```python
# Create a new DataFrame with only the 'Close' values from January 1, 2020, onwards
df1 = df[df.index >= '2020-01-01']['Close']
print(df1)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df1)
plt.title('Stock Price Over Time')
plt.xlabel('Number of Days')
plt.ylabel('Stock Price')
plt.show()
```

```python
# Convert the dataframe to a numpy array
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Reshaping the df1 dataframe to a 2-D array for prediction purpose
scaler=MinMaxScaler(feature_range=(0,1))
df1=scaler.fit_transform(np.array(df1).reshape(-1,1))
print(df1.shape)
```

```python
# Displaying the array
print(df1)
```

```python
from sklearn.model_selection import train_test_split

# Calculate the index for splitting (70% train, 30% test)
split_index = int(0.70 * len(df1))

# Split the array into train and test sets
train_data = df1[:split_index]
test_data = df1[split_index:]

# Print the shapes of the resulting arrays
print("Train data shape:", train_data.shape)
print("Test data shape:", test_data.shape)


```

```python
plt.figure(figsize=(12, 6))
# Plotting the training set
plt.plot(range(len(train_data)), train_data, label='Training Data', color='green')

# Plotting the test set
plt.plot(range(len(train_data), len(df1)), test_data, label='Testing Data', color='red')

# Set title and labels
plt.title('Training and Test Sets')
plt.xlabel('Index')
plt.ylabel('Scaled Values')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Data preprocessing
import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return numpy.array(dataX), numpy.array(dataY)
```

```python
# Assign time steps to look back in the dataset
time_step = 100

# Creating the input and output datasets for time series modeling
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
# reshape input to be [samples, time steps, features] which is required for LSTM
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
# Creating the Stacked LSTM model
model=Sequential()
model.add(LSTM(64,activation = 'relu', return_sequences=True,input_shape=(time_step,1)))
model.add(LSTM(64,return_sequences=True))
model.add(LSTM(64))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=200,batch_size=64,verbose=1)
```

```python
### Lets Do the prediction and check performance metrics
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
train_predict.shape, test_predict.shape
```

```python
##Transformback to original form
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
original_ytrain = scaler.inverse_transform(y_train.reshape(-1,1))
original_ytest = scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# Training Data MSE
mse_train = mean_squared_error(original_ytrain, train_predict)
print("Training Data MSE:", mse_train)

# Test Data MSE
mse_test = mean_squared_error(original_ytest, test_predict)
print("Test Data MSE:", mse_test)
```

```python
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# Evaluation metrices RMSE and MAE
print("Train data RMSE: ", math.sqrt(mean_squared_error(original_ytrain,train_predict)))
print("Test data MAE: ", mean_absolute_error(original_ytrain,train_predict))
print("-------------------------------------------------------------------------------------")
print("Test data RMSE: ", math.sqrt(mean_squared_error(original_ytest,test_predict)))
print("Test data MAE: ", mean_absolute_error(original_ytest,test_predict))
```

```python
print("Train data R2 score:", r2_score(original_ytrain, train_predict))
print("Test data R2 score:", r2_score(original_ytest, test_predict))
```

# **Part D - Testing**


When testing an LSTM model for Stock price prediction, one appropriate method would be to use a holdout dataset. This involves splitting the data into a training set and a test set and using the training set to train the model and the test set to evaluate its performance. The test set should be representative of the entire dataset and should be kept separate until the model has been fully trained.

It is also important to plot the predicted vs actual values for the test set to visualize the performance of the model. This helps to identify any patterns or trends in the data that the model is not capturing and to check if the model is overfitting or underfitting the data.
